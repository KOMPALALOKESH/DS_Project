```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame(data = {"matches win":[25,30,12,15]},index=["ind","eng","aus","pak"])
df
```

```python
# plt.figure(figsize=(20,7)) # also these figure method can change the figsize
df.plot.pie(y="matches win",autopct="%1.1f%%",explode=(0,0,0.1,0),shadow=True,startangle=270,figsize = (8, 8))
plt.title("winning prct of countries below")
```

```python
# pd.set_option('display.max_columns',None)
# pd.set_option('display.max_rows',None)
df = pd.read_csv('crypto_daily_prices.csv')
df
```

```python
df.plot(x='Date',y='BTC',label='BitCoin Price',color='g',linewidth=2,figsize=(15,10))
plt.ylabel('price')
plt.xlabel('date')
plt.title("BitCoin")
plt.legend(loc='upper right')
plt.grid()
```

```python
df.plot(x='Date',y=['BTC','ETH','ADA'],label=['bitcoin','ethercoin','ada'],linewidth=2,xlabel='date',ylabel='prices',title='currencies',figsize=(15,10),grid=True)
plt.legend(loc='upper right')
```

```python
df.plot(x='Date',ylabel='prices',subplots=True,figsize=(15,15),color='ygb',linewidth=2,grid=True)
plt.legend(loc='upper right')
```
